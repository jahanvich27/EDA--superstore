```python
def remove_outlier(df,k=3.33):
  for col in df.columns:
    if df[col].dtype=='int64' or df[col].dtype=='float64':
      mean=df[col].mean()
      global ds
      std=df[col].std()
      out=[i for i in df[col] if i > mean - k*std]
      out=[i for i in out if i < mean + k*std]
      ds=df.loc[df[col].isin(out)]
```

```python
remove_outlier(df,k=3.33)
```
